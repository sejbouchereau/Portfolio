## Overview
Google Fiber provides fiber optic internet to homes and businesses. The customer service team receives calls from customers in different markets.
This project aims to explore **trends in repeat calls** — to help leadership understand how often customers need to call again for the same issue, and why.

**Scenario:**
You are interviewing for a BI analyst position at Google Fiber. As part of the interview process, you have been asked to build a **dashboard** that explores **repeat call trends** and provides insights for improving first-call resolution and customer satisfaction.

  - `Type_5` = Internet & Wi-Fi
- Repeat calls are recorded in seven-day periods.
- Date columns: `contacts_n` = initial contact date; subsequent columns like `contacts_n_6` indicate number of days since first contact (e.g., 6 days after initial contact).
