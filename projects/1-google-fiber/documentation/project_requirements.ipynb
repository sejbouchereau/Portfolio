## 5. User Journeys
### Current User Experience
1. Customer identifies an issue with their service.
2. Customer calls **support**.
3. Agent logs the issue and attempts to resolve it.
4. Customer may need to **call again** if not resolved.
5. Customer experiences delays or frustration.

**Pain Points:**
- Multiple calls for the same issue.
- Limited visibility on repeat calls.
- Lack of insights by **market** or problem type.

---

### Ideal Future User Experience
1. Customer identifies an issue.
2. Customer contacts support, and agent has access to **insights**.
3. Agent resolves the issue on the first call.
4. Customer is satisfied and does not need to call again.
5. Management can monitor trends in repeat calls and improve service.

**Improvements:**
- Reduce repeat calls and wait times.
- Provide data-driven insights for faster resolution.
- Track patterns to prevent recurring issues.
- Increase customer satisfaction and operational efficiency.
